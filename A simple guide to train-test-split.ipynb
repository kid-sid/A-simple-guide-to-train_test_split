```python
from sklearn.datasets import make_blobs               # For creating synthetic data points
from sklearn.model_selection import train_test_split  
```

```python
# create dataset
X, y = make_blobs(n_samples=1000)  # Creating dataset with 1000 samples
#print(X)
#print(y)
print(X.shape)   # X contains 1000 rows and 2 columns
print(y.shape)  
```

```python
# split into train test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33) # Here 33% is the test size and 67% is the train size
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)

X_train2, X_test2, y_train2, y_test2 = train_test_split(X, y, test_size=0.20) # Here 20% is the test size and 80% is the train size
print(X_train2.shape, X_test2.shape, y_train2.shape, y_test2.shape)
```
